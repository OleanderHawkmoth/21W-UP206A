```python
import requests
from pandas.io.json import json_normalize
import pandas as pd

url = "https://api.metro.net/agencies/lametro/routes/2/stops/"
df = pd.read_json(url)
df
```

```python
df = pd.DataFrame(df['items'].tolist())
```

```python
df
```

```python
latitude = df.latitude.mean()
```

```python
longitude = df.longitude.mean()
```

```python
import folium
#initialize map
m = folium.Map(location=[latitude,longitude], tiles='Stamen Terrain', zoom_start=10)
```

```python
# add the stations
for index, row in df.iterrows():
    folium.CircleMarker([row.latitude,row.longitude],
                  radius=5, 
                  popup=row.display_name, 
                  tooltip=row.display_name,
                  fill= True
                 ).add_to(m)
m
```

```python
url = "https://api.metro.net/agencies/lametro/routes/2/vehicles/"
df_live = pd.read_json(url)
df_live
```

```python
df_live = pd.DataFrame(df_live['items'].tolist())
df_live
```

```python
# add the stations
for index, row in df_live.iterrows():
    folium.Marker([row.latitude,row.longitude],icon=folium.Icon(color='red', icon='bus')).add_to(m)
m
```
